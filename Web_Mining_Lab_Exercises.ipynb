```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
#Converting into string
data3=""
for ele in words3:
    data3+=" "
    data3+=ele
       

word_tokens1 = word_tokenize(data3)
```

```python
# Filtering 
STOP_WORDS = nltk.corpus.stopwords.words('english')
terms=[w for w in word_tokens1 if not w in STOP_WORDS]
type(terms)
```

```python
# Frequency of stop words
dic={}
for word in terms:
    if word in dic:
        dic[word]=dic[word]+1
    else:
        dic[word]=1
dic
```

```python
# Finding out total number of terms that are common after removing stop words
len(terms)
```

```python
# Converting into string
data4=""
for ele in terms:
    data4+=" "
    data4+=ele
```

```python
# Writting to index file
text_file = open("index.txt", "w")
n = text_file.write(data4)
text_file.close()
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer 
```

```python
# Initilizing stemmer and Lemmatizer
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer() 
```

```python
# Inputing the index text
file1= open("index.txt", "rt")
data1=file1.read()
words1=data1.split()
file1.close()
stm=[]
lem=[]
words = word_tokenize(data1)
```

```python
# Stemming and Lemmatization
for w in words:
   stm.append(ps.stem(w))
   lem.append(lemmatizer.lemmatize(w))
```

```python
# Length of the stem and lemmatization list
unique_stm = set(stm)
unique_lem = set(lem)
print("Length of unique words from stem list: "+ str(len(unique_stm)))
print("Length of unique words from lemmatization list: "+ str(len(unique_lem)))
```

```python
#Converting stem list to string to writing into file as it has the least count

words5=""
for ele in stm:
    words5+=ele
    words5+=" "
```

```python
# Writing back into file
text_file = open("index.txt", "w")
n = text_file.write(words5)
text_file.close()
```
